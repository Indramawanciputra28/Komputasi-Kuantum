```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

import qiskit
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector, plot_state_qsphere, array_to_latex
```

#1. Qubit dan Keadaanya (Qubit Tunggal)

```python
# Jumlah qubit (n)
qubit = QuantumCircuit(1)
qubit.draw(output='mpl', style='clifford')
```

```python
# Vektor keadaan (Ψ)
Ψ = Statevector(qubit)
Ψ.draw('latex')
```

```python
display(plot_bloch_multivector(qubit))
```

```python
plot_state_qsphere(qubit)
```

```python
# Measurement
qubit.measure_all()

backend = Aer.get_backend('qasm_simulator')
counts = execute(qubit, backend, shots=1024).result().get_counts()
probs = {key: val / 1024 for key, val in counts.items()}

plt.bar(probs.keys(), probs.values(), color='blue', alpha=0.7)
plt.xlabel('States')
plt.ylabel('Probability')
plt.title('Histogram Probabilitas Measurement')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
